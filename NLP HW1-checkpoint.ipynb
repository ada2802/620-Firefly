### Methods

We loaded the meta data and extract customers id for each product. Then we counted the number of reviewers per each product and select the top 100 most reviewed products.  Next, we filter the co-purchasing data (both from and to direction) and meta-data again with these 100 products.

Edge file (product relationship):  The new co purchasing file contains 734 notes (product id) and 900s edges.   In this table, we collects a relation between product A to product B when ether product A or product B are among the top 100 reviewed products.

### Python Code 

```python
id=[]


title=[]

id_str = ""
asin_str = ""
title_str = ""

reviewer=[]

count=0

myfile = open(r"C:\development\data620\project1\amazon-meta.txt", encoding='utf-8' )

output_file = open(r"C:\development\data620\project1\amazon-small-meta.csv", encoding='utf-8',  mode = "w+")


output_file2 = open(r"C:\development\data620\project1\amazon-nodes.csv", encoding='utf-8',  mode = "w+")


output_file3 = open(r"C:\development\data620\project1\amazon-reviewer.csv", encoding='utf-8',  mode = "w+")
     

line = myfile.readline() 
    
while line:

    count+=1

    
    asin_str = ""
    title_str = "" #init value
    group_str = "" #init value
    categories_str = "" #init value
        

    if 'Id:' in line:

        reviewer.clear()
        
        numOfId = line[3:].strip()
        id_str = numOfId #id code as string
    
    
        #start after Id line: read in parallel
    
        line = myfile.readline()
    
     
        if 'ASIN:' in line:
        
            asin_str = line[5:].strip()
    
        line = myfile.readline()
        
        if 'title:' in line:
        
            name=line[8:].strip()
            name = name.replace(',',' ')
                        
            title_str = name

        line = myfile.readline()

        if 'group:' in line:
        
            name=line[8:].strip()
                        
            group_str = name

        
        line = myfile.readline() #read after salesrank
        
        line = myfile.readline() #read after similar
        
        line = myfile.readline() 
        

        if 'categories:' in line:
        
            name=line[14:].strip()
            name = name.replace(',',' ')                        
            categories_str = name


        output_file.writelines((id_str + "," + asin_str + "," + title_str + "," + group_str + "," + categories_str + "\n"))
        output_file2.writelines((id_str + "," + title_str + '\n'))
        
           
              
    
    #skipping line by while loop
    line = myfile.readline()
    if 'cutomer:' in line: #only extract customer line        
        c_values = line.split(':')  
        #['2002-5-13  cutomer',' A2IGOA66Y6O8TQ  rating',' 5  votes','   3  helpful','   2']
                    
        name=c_values[1].replace('rating','')
        name = name.strip()
        reviewer.append(name)
        
        if id_str == '0':
            output_file3.writelines("1," + name + '\n') #unknown bug in first output, use "1,reviewer customer id"
        else:            
            output_file3.writelines(id_str + "," + name + '\n') #id,reviewer customer id
             

    
output_file.close()
output_file2.close()
output_file3.close()

myfile.close()



```
